```python
import numpy as np
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = np.loadtxt("D:\Data Science Projects\Linear regression data science 1\Linear regression on Diabetes Dataset\Traincsv2.csv", delimiter=',')  
x = data[:, :10]
y = data[:, 10]
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.3)

alg1 = LinearRegression()
alg1.fit(x_train, y_train)
TestX=np.loadtxt("D:\Data Science Projects\Linear regression data science 1\Linear regression on Diabetes Dataset\TestCsv1.csv", delimiter=',')
prediction = alg1.predict(TestX)
np.savetxt("prediction.csv", prediction, delimiter=",",fmt='%.5f')
y_predicted=alg1.predict(x_test)
plt.scatter(y_test,y_predicted)
```
